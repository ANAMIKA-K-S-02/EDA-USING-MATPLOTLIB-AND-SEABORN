```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("football.csv")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Nationality",data=dataset,order=dataset["Nationality"].value_counts().index[:10])
plt.title("TOP 10 NATIONALITIES BASED ON PLAYER COUNT.")
plt.ylabel("PLAYER COUNT")
plt.xlabel("NATIONALITY")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(dataset["Age"],bins=15,kde=True,color="red")
plt.title("PLAYERS BASED ON AGE")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=dataset.iloc[:30,dataset.columns.get_loc("Age")],y=dataset.iloc[:30,dataset.columns.get_loc("Wage")],marker="o",color="red")
plt.title("Age vs Wage map")
plt.xlabel("Age")
plt.ylabel("Wage")
plt.grid(True)
plt.show()
```

```python
top_nation=dataset["Nationality"].value_counts().index[:5]
plt.figure(figsize=(10,6))
sns.boxplot(x="Nationality",y="Wage",data=dataset[dataset["Nationality"].isin(top_nation)],palette="Set3")
plt.title("Wage Distribution by Nationality")
plt.xlabel("Nationality")
plt.ylabel("Wage")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Club",data=dataset,order=dataset["Club"].value_counts().index[:5],palette="muted")
plt.title("Top 5 Clubs by Player Count")
plt.xlabel("Club")
plt.ylabel("Player Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_10=dataset.nlargest(10,"Value")[['Name', 'Value']]
plt.figure(figsize=(10,6))
sns.barplot(x="Value",y="Name",data=top_10,palette="coolwarm")
plt.title("Top 10 Players by Value")
plt.xlabel("Value")
plt.ylabel("Player Name")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(dataset["Age"],bins=15,kde=True,color="green")
plt.title("Distribution of Player Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Age",y="Wage",data=dataset,color="blue")
plt.title("Scatter Plot:Age vs Wage")
plt.xlabel("Age")
plt.ylabel("Wage")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Position",y="Wage",data=dataset,palette="Set2")
plt.title("EAGE DISTRIBUTION BY POSITION")
plt.xlabel("Position")
plt.ylabel("Wage")
plt.xticks(rotation=45)
plt.show()
```

```python
top_20=dataset.nlargest(20,"Value")[['Age','Value']]
plt.figure(figsize=(10,6))
sns.lineplot(x='Age',y='Value',data=top_20,marker="o",color="purple")
plt.title("Age vs Value")
plt.xlabel("Age")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
skills=['Crossing', 'Finishing', 'ShortPassing', 'Dribbling', 'BallControl', 'ShotPower', 'Stamina', 'LongShots']
plt.figure(figsize=(10,6))
sns.heatmap(dataset[skills].corr(),annot=True,cmap="viridis",fmt="0.2f")
plt.title("Correlation Between Technical Skills")
plt.show()
```

```python
top_10_clubs=dataset['Club'].value_counts().head(10)
plt.figure(figsize=(12,8))
sns.barplot(x=top_10_clubs.values,y=top_10_clubs.index,palette="muted")
plt.title("Top 10 Clubs by Number of Players")
plt.xlabel("Number of Players")
plt.ylabel("Club")
plt.show()
```

```python
plt.figure()
sns.pairplot(dataset[['Age','Wage','Value']])
plt.suptitle("Pair Plot: Age, Wage and value", y=1.02)
plt.show()

```

```python
top_nations=dataset['Nationality'].value_counts().head(5).index
plt.figure(figsize=(10, 6))
sns.boxplot(x='Nationality', y='Strength', data=dataset[dataset['Nationality'].isin(top_nations)], palette="pastel")
plt.title("Strength Distribution by Top 5 Nationalities")
plt.xlabel("Nationality")
plt.ylabel("Strength")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Dribbling',y="BallControl",data=dataset,hue='Position',palette="tab10")
plt.xlabel("Dribbling")
plt.ylabel("Ball Control")
plt.legend(loc="upper left",bbox_to_anchor=(1,1))
plt.show()
```
